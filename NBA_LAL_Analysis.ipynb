```python
#Imported necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Imported the dataset containing all the players of Los Angeles Lakers and assists done by them 
dataset_LAL = pd.read_csv('/Users/jaynilvora/Downloads/Ast_LAL.csv', encoding='latin')
dataset_LAL.head()
```

I wanted to analyze players in Los Angeles Lakers who are really good at creating plays by passing. There players who don't get a ton of traditional assists, but get a lots of "Secondary Assists". The guy who passes it to the assister. The hockey assist.

```python
#Scatterplot shows all the players and their Primary and Secondary Assists 
#of Los Angeles Lakers
plt1 = dataset_LAL.plot(kind='scatter',x='SECONDARY AST',y='AST',color="Red",alpha=0.6) 
plt1.text(14,89,"Julius Randle",size='large')
plt1.set_xlabel('Secondary Assists')
plt1.set_ylabel('Primary Assists')
```

I plotted the Primary assists and Secondary assists of every player. 
As we can see that Julius Randle has the most number of secondary assists.

```python
#Imported the dataset containing players of all NBA teams and assists done by them
dataset_allTeams = pd.read_csv('/Users/jaynilvora/Downloads/AST.csv', encoding='latin')
dataset_allTeams.head()
```

```python
#Counted the total secondary assists done by all teams
#Grouped by teams
#Bar graph to show secondary assists by all teams

team_AST = dataset_allTeams.groupby('TEAM').sum()
plt2 = team_AST['SECONDARY AST'].sort_values(ascending=False).plot(kind='bar')
plt2.set_ylabel('Secondary Assists')
plt2.set_xlabel('Teams')
plt2.set_title('Secondary assists by team')
```

As we can see from the bar graph above, Los Angeles Lakers is not very good at secondary assists. It stands at 26th position. This is an area where the team can immprove.

Analysis of drives of Los Angeles Lakers.
Driving to the basket is standard in the NBA. You can make a layup, get fouled, or kick it out for a 3PT assist. I was curious how the team did compared to other teams. I wanted to analyze the drives of Los Angeles Lakers.

```python
#Imported the dataset containing players of all teams and information on their drives
Drives_All = pd.read_csv('/Users/jaynilvora/Downloads/drives.csv', encoding='latin')
team = Drives_All.groupby('TEAM').sum()
Drives_All.head()
```

```python
#Imported the dataset containing players of Los Angeles Lakers and information on their drives
Drives_LAL = pd.read_csv('/Users/jaynilvora/Downloads/Drives_LAL.csv', encoding='latin')
Drives_LAL.head()
```

```python
#Calculated points per drive
team['PTSperDRIVE'] = team['PTS']/team['DRIVES']
```

```python
#Bar graph to show Points per Drive of all teams
plt3 = team['PTSperDRIVE'].sort_values(ascending=False).plot(kind='bar')
plt3.set_ylabel('Points per Drive')
plt3.set_xlabel('Teams')
plt3.set_title('Points per drive by team')

#Bar graph to show total number of drives by each player in Miami HEAT
plt4 = Drives_LAL.plot(kind='bar',x='PLAYER',y='DRIVES',color="Red",alpha=0.6) 
plt4.set_ylabel('Drives')
plt4.set_xlabel('Players')
plt4.set_title('Total drives for Los Angeles Lakers players')
```

It turned out the team does well in Points per Drive as compared to other teams in NBA. Jordan Clarkson has the most number of drives.
